```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from pymongo import MongoClient
from pymongo.errors import ServerSelectionTimeoutError
import holidays
```

```python
# MongoDB connection URI
mongo_uri = 'mongodb://localhost:27017/'  # Replace with your MongoDB URI

# Database and collection names
db_name = 'metro_data'  # Replace with your database name
collection_name = 'demo_data'  # Replace with your collection name

try:
    # Establish connection to MongoDB
    client = MongoClient(mongo_uri, serverSelectionTimeoutMS=5000)  # Timeout for connection
    print("Connected to MongoDB!")

    # Access the database and collection
    db = client[db_name]
    collection = db[collection_name]
    print(f"Accessing collection: {collection_name}")

    # Fetch all documents from the collection
    cursor = collection.find()

    # Convert the cursor to a list of dictionaries
    documents = list(cursor)

    # Convert the list of dictionaries to a pandas DataFrame
    data = pd.DataFrame(documents)
    print("Data converted to DataFrame.")

    # Display the head of the DataFrame
    print("Head of the DataFrame:")
    print(data.head())

except ServerSelectionTimeoutError as e:
    print("Failed to connect to MongoDB:", e)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Scatter plot to check linearity between 'Dt_encoded' and '_id_encoded'
plt.figure(figsize=(8,6))
plt.scatter(data['Dt'], data['_id'], alpha=0.5, color='blue')

# Adding titles and labels
plt.title('Linearity Check between Dt_encoded and _id_encoded')
plt.xlabel('Dt_encoded')
plt.ylabel('_id_encoded')

# Rotate x-axis labels if necessary (especially for dates)
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
data['Dt'] = pd.to_datetime(data['Dt'], errors='coerce')
data['hour'] = data['Dt'].dt.hour
data['day_of_week'] = data['Dt'].dt.day_name()
data['date'] = data['Dt'].dt.date
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='hour', data=data)
plt.title('Footfall by Hour')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='day_of_week', data=data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Footfall by Day of the Week')
plt.show()

# Use the 'holidays' library to get public holiday dates for India
indian_holidays = holidays.India()

# Create a new column indicating whether the date is a holiday
data['holiday_name'] = data['date'].apply(lambda x: indian_holidays.get(x) if x in indian_holidays else None)

# Filter only the rows that correspond to holidays
holiday_footfall = data[data['holiday_name'].notnull()]

# Plot footfall on specific Indian holiday dates
plt.figure(figsize=(12, 6))
sns.countplot(x='holiday_name', data=holiday_footfall)
plt.title('Footfall on Specific Indian Holidays')
plt.xticks(rotation=45, ha='right')  # Rotate the holiday names for better readability
plt.tight_layout()
plt.show()
```
